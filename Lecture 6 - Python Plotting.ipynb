### Lambda Functions in Python
A `lambda` function in Python is an anonymous, single-expression function. It is defined using the `lambda` keyword and allows for the creation of small, inline functions without needing to use the def keyword or assign a name. `lambda` functions are particularly useful in situations where a small, one-off function is needed as an argument to a higher-order function (a function that takes other functions as arguments). We will discuss more in future lectures. 

```python
square = lambda x: x**2 
print(square(5))  # Output: 25
```

```python
# Sort a dictionary using only the built-in sorted() function.
# Show the original dictionary, the sorted keys, and items sorted by value.

items = {'Milk': 4.5, 'Bread': 3.2, 'Eggs': 6.0, 'Cheese': 8.5, 'Fruits': 12.0, 'Vegetables': 10.0, 'Snacks': 9.0, 'Beverages': 7.5}

# Print the original dictionary
print("Original items (dict):", items)

# sorted() on a dict returns a sorted list of keys by default
sorted_keys = sorted(items)
print("Sorted keys:", sorted_keys)

# To sort by values, use sorted() on items() with a key function (still using only sorted())
sorted_by_value = sorted(items.items(), key=lambda kv: kv[1]) # Always ascending. Need to use lambda function. 
sorted_by_value_decending = sorted(items.items(), key= lambda kv:kv[1], reverse = True) # Need to add reverse = True
print("Items sorted by value (ascending):", sorted_by_value)
print("Items sorted by value (decending):", sorted_by_value_decending)
```

## Use NumPy to creating arrays: `array`, `linspace`, `arange`, helpers

```python
# Please note that I intentionally did not add comments. 
# Add comments to code in this notebook.

import numpy as np # Importing numpy module

# 1D arrays
a = np.array([1.2, 3.5, 5.4, 7.1], dtype=float) # Defining array and setting it as type
x_lin = np.linspace(0.0, 5.0, 6)     # Returns evenly spaced integers over a specified interval -- inclusive ends
x_rng = np.arange(0.0, 5.0, 1.5)      # half-open [start, end)

# Note to self: linspace works in terms of choosing the number of integers. (start, stop, number of integers) step counter will be altered to ensure that the desired number of integers is produced
# Meanwhile, arange works in terms of choosing a definitive step size. (start, stop, step size) step size works independepently

print('a      =', a)  # Prints the array variable defined
print('lin    =', x_lin) # Prints the linspace that was defined above
print('arange =', x_rng) # Displays the defined x

# 2D helpers
Z = np.zeros((5,2))  # Makes an array of 0's according to the integers given
O = np.ones((4,6))  # Makes an array of 1's according to the integers given
I = np.eye(2)  # Square matrix that is whatever is given; Diagonal of 1's (think of RREF)
print('zeros:') 
print(Z)  # Prints the variable Z
print('ones:')  
print(O)  # Prints the variable O
print('eye:')
print(I)  # Print the variable I

print(a[3])
```

## Basic 2D Plotting

```python
# Try comment out the lines of code below and run it to see the plot
# So you can see how the plot changes
# Add comments to explain what each line does

import matplotlib.pyplot as plt  
import numpy as np   # Importing matplotlib and numpy modules
x = np.linspace(-10, 10, 100) # Change the range and number of points to see what happens
y = x**2    # Setting y as a function of x
plt.plot(x, y)  # Plotting the variables as a plot 
plt.title("Parabola: $y=x^2$")   # Naming the plot 
plt.xlabel("x-axis")
plt.ylabel("y-axis")  # Naming x and y axis
plt.grid()  # Adds gridlines to graph
# plt.xlim(0, 5)  # Uncomment to limit x-axis from 0 to 5
# plt.ylim(0, 25) # Uncomment to limit y-axis from 0 to 25      # These lines adjust the viewing window and adjusts the section of the parabola that we see
plt.axhline(0, color='green',linewidth=1.5, ls='dotted')              # Adds horizontal line to the indicated position; color, line width, and line type is customizable
plt.show()
```

## Multiple 2D Plot Types

```python
# Try comment out the lines of code below and run it to see the plot
# So you can see how the plot changes
# Add comments to explain what each line does


import matplotlib.pyplot as plt
import numpy as np               # Importing matplot and numpy modules


x = np.arange(0, 11)   # Producing a range of integers from 0 - 11 in evenly spaced steps
# y= x**2              # Alternative way to define y as a function of x
y = np.power(x, 2)     # Setting y as a function of x

plt.figure(figsize=(14, 8))
plt.subplot(2, 3, 1); plt.plot(x, y); plt.title('Plot'); plt.grid()
plt.subplot(2, 3, 2); plt.scatter(x, y); plt.title('Scatter'); plt.grid()
plt.subplot(2, 3, 3); plt.bar(x, y); plt.title('Bar'); plt.grid()
plt.subplot(2, 3, 4); plt.loglog(x, y); plt.title('Loglog'); plt.grid()
plt.subplot(2, 3, 5); plt.semilogx(x, y); plt.title('Semilogx'); plt.grid()
plt.subplot(2, 3, 6); plt.semilogy(x, y); plt.title('Semilogy'); plt.grid()
plt.tight_layout()
plt.show()
```

## 3D Plotting Examples
This section is optional. Proceed if you are interested.

```python
# Create a 3D parametric plot for x=cos(t),y=sin(t),z=t
import numpy as np
import matplotlib.pyplot as plt


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
t = np.linspace(0, 10*np.pi, 500)
x = np.sin(t)
y = np.cos(t)
ax.plot3D(x, y, t)
ax.set_title("3D Parametric Plot")
plt.show()
```

```python
# 3D surface plot of Z=exp(X^2Y^2)- Gaussian Function
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create grid
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute Z
Z = np.exp(-X**2 - Y**2) # Z=exp(X^2Y^2)- Gaussian Function

# Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Labels and title
ax.set_title('z = exp(-x - y))')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Colorbar
fig.colorbar(surf, shrink=0.5, aspect=10)

plt.show()
```
